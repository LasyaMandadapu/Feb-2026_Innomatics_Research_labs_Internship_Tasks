Problem-1: Mobile Recharge Validation

```python
# Problem 1: Mobile Recharge Validation

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    print("Recharge Amount Entered:", amount)

    if amount < 50:
        print("Result: Recharge must be at least ₹50")
    elif amount not in valid_plans:
        print("Result: Invalid Plan")
    else:
        print("Result: Recharge Successful")


# Sample input
validate_recharge(199)
```

Problem-2: Inventory Reorder Alert

```python
# Problem 2: Inventory Reorder Alert

def check_inventory(products):
    print("Product Stock Data:", products)

    for item, stock in products.items():
        if stock < 15:
            print(item, "→ Reorder Alert")
        else:
            print(item, "→ Stock OK")


products = {"Pen": 10, "Notebook": 25, "Marker": 12}
check_inventory(products)
```

Problem-3: Student Result Processing

```python
# Problem 3: Student Result Processing

def student_result(marks):
    print("Marks:", marks)

    avg = sum(marks) / len(marks)
    print("Average:", avg)

    if avg >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")


student_result([60, 45, 70, 50])
```

Problem-4: Cab Fare Estimator

```python
# Problem 4: Cab Fare Estimator

def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)

    print("Distance:", distance)
    print("Peak Hour:", peak)

    if peak == "yes":
        fare *= 1.25

    print("Total Fare:", fare)


calculate_fare(10, "yes")
```

Problem-5: Attendance Eligibility

```python
# Problem 5: Attendance Eligibility Checker

def check_eligibility(attendance):
    print("Attendance:", attendance)

    percent = (attendance.count("P") / len(attendance)) * 100
    print("Percentage:", percent)

    if percent >= 75:
        print("Status: Eligible")
    else:
        print("Status: Not Eligible")


check_eligibility(["P", "P", "A", "P", "P"])
```

Problem-6: Password Strength Checker

```python
# Problem 6: Password Strength Checker

def check_password(password):
    print("Password Entered:", password)

    has_digit = any(ch.isdigit() for ch in password)
    has_special = any(ch in "@#$" for ch in password)

    if len(password) < 8:
        print("Strength: Weak Password")
    elif has_digit and has_special:
        print("Strength: Strong Password")
    else:
        print("Strength: Weak Password")


check_password("Abc@1234")
```
