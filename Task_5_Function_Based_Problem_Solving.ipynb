Problem-1: Social Media- Post Engagement Analyzer

```python
# Social Media – Post Engagement Analyzer

def analyze_post_engagement(likes_list):

    print("Likes Received Per Hour:", likes_list)

    # Calculate total likes using loop
    total_likes = 0
    for like in likes_list:
        total_likes += like

    print("Total Likes:", total_likes)

    # Check viral condition
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Sample Input (visible in output)
likes_data = [200, 150, 500, 300, 100]

analyze_post_engagement(likes_data)
```

Problem-3: Agriculture-Rainfall Adequacy Checker

```python
def check_rainfall(rainfall_data, required_level):

    print("Seasonal Rainfall Data:", rainfall_data)

    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    print("Average Rainfall:", average)

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Sample Input
rain_data = [70, 75, 100, 65]
check_rainfall(rain_data, 70)
```

Problem-4: Social Media- Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):

    print("Usernames List:", usernames)

    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users = ["user1", "user1", "user3", "user4"]
detect_duplicate_accounts(users)
```

Problem-6: Agriculture- Premium Crop Price Filter

```python
def filter_premium_crops(prices):
    print("Crop Prices:", prices)

    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)
filter_premium_crops([2500, 3500, 1700, 3200])
```

Problem-9: E-Learning- Attendance Eligibility System

```python
def check_exam_eligibility(attendance):
    print("Attendance Record:", attendance)

    present = 0
    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
check_exam_eligibility(["P", "P", "A", "P", "P"])
```

Problem-10: Smart Electricty Bill Analyzer

```python
def calculate_electricity_bill(units):
    print("Units Consumed:", units)

    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    print("Total Bill:", bill)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    print("Usage Classification:", status)

calculate_electricity_bill(250)
```
