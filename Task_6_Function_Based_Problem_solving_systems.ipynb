 ** problem-1: Smart Parking Lot Management System **

```python
def manage_parking(capacity, vehicle_logs):
    print("Parking Capacity:", capacity)
    print("Vehicle Logs:", vehicle_logs)

    parked = 0
    peak = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full - Alert!")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)
manage_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

 ** Problem-2: Online Food Delivery Time estimator **

```python
def estimate_delivery_time(distance, traffic, weather):
    print("Distance (km):", distance)
    print("Traffic Level:", traffic)
    print("Weather:", weather)
    base_time = distance * 5  # 5 minutes per km
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 8

    if weather == "Rainy":
        base_time += 10

    print("Estimated Delivery Time:", base_time, "minutes")
estimate_delivery_time(8, "High", "Rainy")
```

** Problem-3: Movie Theatre seat Occupancy Analyzer **

```python
def analyze_occupancy(total_seats, booked_seats):
    print("Total Seats:", total_seats)
    print("Booked Seats Count:", len(booked_seats))
    occupancy = (len(booked_seats) / total_seats) * 100
    print("Occupancy:", int(occupancy), "%")
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
analyze_occupancy(200, [1]*150)
```

** Problem-4: Cloud Server Load Classification **

```python
def classify_server_load(cpu_readings):
    print("CPU Readings (%):", cpu_readings)

    avg = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
classify_server_load([45, 60, 70, 85, 90])
```

** Problem-5: Smart Classroom Resource Usage Monitor **

```python
def monitor_resources(resource_usage):
    print("Resource Usage Data:", resource_usage)

    overused = []
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
monitor_resources({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

** Problem-6: Online Event Registration Capacity Controller **

```python
def manage_registrations(capacity, registrations):
    print("Event Capacity:", capacity)
    print("Total Registrations:", registrations)

    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
manage_registrations(100, 105)
```
